**Rastrigin function**

image.png

```python
import math

def Rastrigin_function(x):
  n = len(x)
  sum = 0.
  A = 10
  for i in range(0,n):
    if (-5.12 <= x[i] and x[i] <= 5.12):
      sum += x[i] ** 2 - A*(math.cos(2 * math.pi * x[i]))
    else : 
      return "-5.12 <= x[i] <= 5.12 조건에 맞게 x[i]를 변경하세요."    
  return (A + sum)

Rastrigin_function([1.3, 5.1, 2.4])
```

Sphere function
image.png

```python
def sphere_function(x):
  a = 0
  n = len(x)
  for i in range(0, n):
    a += x[i] ** 2
    
  return a

output = sphere_function([10,5,2,8])
output
```

**Rosenbrock function**
image.png

```python
def Rosenbrock_function(x):
  n = len(x) - 1
  sum = 0
  if (n+1 >= 2):
    for i in range(0,n):
      sum += 100 * (x[i+1] - (x[i] ** 2)) ** 2 + (1 - (x[i] ** 2))
  else : 
    return "n의 개수를 2 이상으로 변경하세요."
  return sum

Rosenbrock_function([3])
```
